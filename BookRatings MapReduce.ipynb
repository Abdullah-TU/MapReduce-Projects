```python
%%writefile ratings.py

from mrjob.job import MRJob
from mrjob.protocol import RawValueProtocol
from mrjob.step import MRStep

class ratings(MRJob):
    OUTPUT_PROTOCOL = RawValueProtocol
    def mapper(self, _, line):
        counts=line.replace('"', '').split(';')
        word=counts[1].split(',')
        try:
            if len(word)==3 and len(word[2].strip())>2 and str(word[2].strip())!="n/a" and str(word[2].strip())!="\\n/a\\":
                country, userID= word[2].strip(), int(counts[0])
                yield userID, country
            else:
                rating, userID = int(counts[2]), int(counts[0])
                yield userID, rating
        except:
            pass
            
    def reducer_step_1(self, key, values):
        data=[]
        for m in values:
            data.append(m)
        if len(data)>1:
            data=list(reversed(data))
            if isinstance(data[0], str):
                try:
                    yield data[0],list(map(int, data[1:]))   
                except:
                    pass
    def reducer_step_2(self, key, values):
        countx=0
        sumx=0
        for m in values:
            countx+=len(m)
            sumx+=sum(m)
        yield None, '{};{};{:.1f}'.format(key, countx, sumx/countx)
    
    def steps(self):
        return [MRStep(mapper=self.mapper, reducer=self.reducer_step_1), MRStep(reducer=self.reducer_step_2)]
if __name__=="__main__":
    ratings.run()
```
